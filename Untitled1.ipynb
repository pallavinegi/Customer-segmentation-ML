 Importing the essential Libraries for Data Analysis and Clustering


```python
# ðŸ“¦ Importing Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

this is the step for loading the dataset(mall customer data)

```python
df = pd.read_csv('/content/Mall_Customers.csv')  # Change path if running locally
df.head()
```

BAsic EDA operations to check shape,info,and missing values

```python
#EDA

# Shape & Info
print("Shape:", df.shape)
df.info()
```

```python
# Check missing values
print("\nMissing values:\n", df.isnull().sum())

```

to visualise the gender distribution in the dataset

```python
# Gender Count
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

to visualise the age distribution using histogram

```python
# Age Distribution
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.show()
```

 Scatter plot to understand the relation between Income & Spending Score


```python
# Income vs Spending Score
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender', data=df)
plt.title("Income vs Spending Score by Gender")
plt.show()
```

Feature Selection - Selecting Income & Spending Score for Clustering


```python
# FEATURE SELECTION AND SCALING


# Using only Annual Income & Spending Score for clustering
x = df.iloc[:, [3, 4]].values
```

Feature Selection - Selecting Income & Spending Score for Clustering


```python
# Scaling
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

Finding the optimal number of clusters using Elbow Method (WCSS curve)


```python
#Elbow method to find optimal cluster


wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

Applying KMeans Clustering with k=5


```python
#KMEANS CLUSTRING(USING K=5)


kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y = kmeans.fit_predict(x_scaled)

# Add cluster to dataframe
df['Cluster'] = y

```

 Adding cluster labels to original dataframe


```python
#Visualize the cluster


plt.figure(figsize=(8, 6))
colors = ['green', 'red', 'blue', 'yellow', 'violet']

```

 Visualizing the Clusters and their Centroids


```python
for i in range(5):
    plt.scatter(x_scaled[y == i, 0], x_scaled[y == i, 1], s=50, c=colors[i], label=f'Cluster {i+1}')
```

```python
# Centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='black', marker='X', label='Centroids')

plt.title('Customer Clusters (KMeans)')
plt.xlabel('Annual Income (scaled)')
plt.ylabel('Spending Score (scaled)')
plt.legend()
plt.grid(True)
plt.show()
```

Cluster Profile Summary - Understanding average features per cluster


```python
# CLUSTER PROFILE SUMMERY


# Drop Gender for numeric mean summary
cluster_summary = df.drop('CustomerID', axis=1).groupby('Cluster').mean(numeric_only=True)
print(cluster_summary)
```

PCA (Principal Component Analysis) is used here to reduce the data dimensions from 2D+ to 2D
# for easier visualization of clusters while preserving the maximum variance (important information) in the data.


```python
#PCA FOR BETTER VIZUALIZATION
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)

plt.scatter(x_pca[:, 0], x_pca[:, 1], c=y, cmap='rainbow')
plt.title("Clusters visualized using PCA")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.show()

```
